I want to visualise the function: **Z = X^2 + Y^2**

```python
x = np.linspace(-5,5,11)
y = np.linspace(-5,5,11)
X,Y = np.meshgrid(x,y)
Z = X**2+Y**2 # Element wise operation 
```

Same if i wrote Z[ i , j ] = X[ i , j ]**2 + Y[ i , j ]**2

```python
Z
```

```python
plt.scatter(Y,Z)
```

```python
# we can see the preliminary convex shape of this plot
```

```python
fig=plt.figure()
plt.contour(X,Y,Z,cmap='plasma')
plt.axis('scaled')
plt.show()
plt.savefig('contour.png')
```

The Z meshgrid is consisted of every point on X and Y grid

```python
fig=plt.figure()
```

```python
fig=plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma',alpha=.9)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.savefig('3D surface plot')
 
```

```python
fig=plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma',alpha=.9)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(1,60)
plt.savefig('3D surface plot')
 
```
